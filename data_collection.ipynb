```python
ma = pd.read_csv(r"C:\Users\hp\Downloads\MAC corpus.csv", sep=',', encoding='utf-8')
ma.head()
```

```python
# Renommer les colonnes de ma
ma = ma.rename(columns={'tweets': 'text'})

# Afficher le résultat
print(ma)
```

```python
# Lecture des fichiers CSV
import csv
t1 = pd.read_csv(
    r"C:\Users\hp\Downloads\train_pos.csv",
    encoding='utf-8',          
    header=None,               # pas d'en-tête
    quoting=csv.QUOTE_NONE,    # ne pas interpréter les guillemets
    on_bad_lines='skip'        # ignorer les lignes corrompues si elles existent
)


t2 = pd.read_csv(
    r"C:\Users\hp\Downloads\train_neg.csv",
    encoding='utf-8',
    header=None,
    quoting=csv.QUOTE_NONE,
    on_bad_lines='skip'
)

t3 = pd.read_csv(
    r"C:\Users\hp\Downloads\test_pos.csv",
    encoding='utf-8',
    header=None,
    quoting=csv.QUOTE_NONE,
    on_bad_lines='skip'
)


t4 = pd.read_csv(
    r"C:\Users\hp\Downloads\test_neg.csv",
    encoding='utf-8',
    header=None,
    quoting=csv.QUOTE_NONE,
    on_bad_lines='skip'
)

```

```python
# Concaténer les DataFrames
TN= pd.concat([t1, t2, t3, t4], ignore_index=True)
TN.shape
```
